```python
import os
import shutil
import cv2
from pprint import pprint
```

```python
# data_dir = '/tf/mount/Dataset/mimed/Dataset_15_09'
data_dir = '/tf/mount/Dataset_15_09'
dir_list = os.listdir(data_dir)
```

# used for changing the file name into 6 digit for dataset 15_09

```python
main_counter = 0
for f_dir in dir_list:
    img_dir = os.path.join(data_dir, f_dir,"Bilder")
    img_list = os.listdir(img_dir)
    for i in range (1, len(img_list)+1):
        if f_dir == 'JPEGImages':
            img_path = os.path.join(img_dir,str(i).zfill(6)+'.jpg')
        else:
            img_path = os.path.join(img_dir,str(i)+'.jpg')
        os.rename(img_path,'/tf/project/dataset/'+ str(main_counter).zfill(6)+'.jpg')
        main_counter += 1
```

```python
img_dir = '/tf/project/dataset/JPEGImages'
img_list = os.listdir(img_dir)
for img in img_list:
    img_path = os.path.join(img_dir, img)
    f_img = cv2.imread(img_path)
    shape = f_img.shape
    h = shape[0]
    w = shape[1]
    if h == 1080:
        f_img = f_img[25:1055, 150:1470]
        cv2.imwrite(img_path,f_img)
```

# used for finding test and train set from json, sorting the jpeg file

```python
anno_dir = os.path.join(data_dir,'Annotations')
img_dir = os.path.join(data_dir,'JPEGImages')
```

```python
anno_list = os.listdir(anno_dir)
img_list = os.listdir(img_dir)
```

```python
import random
random.shuffle(anno_list)
sum_num = len(anno_list) 
train_num = int(0.8 * sum_num)
test_num = sum_num - train_num

train_list = anno_list[:train_num]
test_list = anno_list[train_num:]

def check_folder_exist(folder):
    if not os.path.exists(folder):   # check the path
        os.makedirs(folder)          # if not exis, create
```

```python
pprint(test_list)
```

```python
for j in train_list:
    name = j.split('.')[0]
    before_img_path = os.path.join(img_dir,name+'.png')
    after_img_path = os.path.join(img_dir,'train')
    check_folder_exist(after_img_path)
    after_img_path = os.path.join(after_img_path, name+'.png')
    
    os.rename(before_img_path, after_img_path)
    
    before_anno_path = os.path.join(anno_dir, j)
    after_anno_path = os.path.join(anno_dir,'train')
    check_folder_exist(after_anno_path)
    after_anno_path = os.path.join(after_anno_path, j)
    
    os.rename(before_anno_path, after_anno_path)
    
for j in test_list:
    name = j.split('.')[0]
    before_img_path = os.path.join(img_dir,name+'.png')
    after_img_path = os.path.join(img_dir,'test')
    check_folder_exist(after_img_path)
    after_img_path = os.path.join(after_img_path, name+'.png')

    os.rename(before_img_path, after_img_path)
    
    before_anno_path = os.path.join(anno_dir, j)
    after_anno_path = os.path.join(anno_dir,'test')
    check_folder_exist(after_anno_path)
    after_anno_path = os.path.join(after_anno_path, j)
    
    os.rename(before_anno_path, after_anno_path)
```
